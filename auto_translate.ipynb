```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
import re

# Load tokenizer & model
model_name = "facebook/nllb-200-3.3B"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# ✅ Danh sách ngôn ngữ có sẵn
language_options = {
    "1": ("English", "eng_Latn"),
    "2": ("French", "fra_Latn"),
    "3": ("German", "deu_Latn"),
    "4": ("Chinese (Simplified)", "zho_Hans"),
    "5": ("Chinese (Traditional)", "zho_Hant"),
    "6": ("Japanese", "jpn_Jpan"),
    "7": ("Korean", "kor_Hang"),
    "8": ("Russian", "rus_Cyrl"),
    "9": ("Spanish", "spa_Latn"),
}

# Hàm chia nhỏ văn bản thành câu
def split_sentences(text):
    sentences = re.split(r'([.!?])', text)
    result = []
    for i in range(0, len(sentences)-1, 2):
        result.append(sentences[i].strip() + sentences[i+1].strip())
    return result

# Hàm dịch từng câu
def translate(text, src_lang_id, tgt_lang_id):
    inputs = tokenizer(text, return_tensors="pt")
    output_ids = model.generate(
        **inputs,
        forced_bos_token_id=tgt_lang_id,  # ✅ Dùng ID đã chọn sẵn
        max_length=200,
        do_sample=False
    )
    return tokenizer.decode(output_ids[0], skip_special_tokens=True)
```

```python
# --- Giao diện chọn ngôn ngữ ---
print("\n🌍 Chọn ngôn ngữ bạn muốn nhập:")
for key, (lang_name, _) in language_options.items():
    print(f"{key}. {lang_name}")

selected_lang = input("\nNhập số tương ứng với ngôn ngữ của bạn: ")

if selected_lang not in language_options:
    print("❌ Lựa chọn không hợp lệ. Mặc định là English.")
    src_lang_name, src_lang_code = "English", "eng_Latn"
else:
    src_lang_name, src_lang_code = language_options[selected_lang]
    print(f"✅ Bạn đã chọn: {src_lang_name}")

# ✅ Lấy ID của ngôn ngữ đã chọn
src_lang_id = tokenizer.convert_tokens_to_ids(src_lang_code)
tgt_lang_id = tokenizer.convert_tokens_to_ids("vie_Latn")  # Mặc định dịch sang tiếng Việt

# --- Nhập văn bản và dịch ---
while True:
    text = input("\nNhập văn bản (hoặc gõ 'exit' để thoát): ")
    if text.lower() == "exit":
        break

    print("🔄 Đang dịch...")

    sentences = split_sentences(text)
    translated_sentences = [translate(sentence, src_lang_id, tgt_lang_id) for sentence in sentences]

    print("📜 Bản dịch:", " ".join(translated_sentences))
```
